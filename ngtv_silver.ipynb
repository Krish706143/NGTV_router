```python
bronze_df1 = spark.read.format("delta").load(
    "abfss://bronze@ngtv.dfs.core.windows.net/df_49_50_error_disable_logs_merge"
)

bronze_df1.show(100, truncate=False)
```

```python
bronze_df1.printSchema()
```

```python
# removing dupplicates
removeduplicate_df1 = bronze_df1.dropDuplicates(["device"])
```

```python
removeduplicate_df1.printSchema()
```

```python
removeduplicate_df1.show(100, truncate=True)
```

```python
silver_path = "abfss://silver@ngtv.dfs.core.windows.net/"
```

```python
removeduplicate_df1.write.format("delta").mode("overwrite").save(
    silver_path + "df_49_50_error_disable_logs_merge"
)
```
